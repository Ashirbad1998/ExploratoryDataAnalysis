"Quantity" can't be in negative format we have to treat the "Quantity" feature and make it positive

```python
df1['Year'] = df1['InvoiceDate'].dt.year
```

### 4) Data Visualization

```python
df1.Year.value_counts().plot.bar()
plt.show()
```

**Insights:**  "Sales" are much more higher in 2011 year than 2010 year.

```python
round(df1[df1.Year==2010].groupby("Country")["Sales"].sum().head(20).sort_values(ascending=False),0).plot.bar()
```

**Insights:** Germany have maximum Sales in the year 2010.

```python
df1[df1.Year==2011]
```

```python
round(df1[df1.Year==2011].groupby("Country")["Sales"].sum().head(20).sort_values(ascending=False),0).plot.bar()
```

**Insights:** EIRE took 1st position and Germany got 2nd position in sales in the year 2011.

```python
round(df1[df1.Year==2010].groupby("Description")["Sales"].sum().sort_values(ascending=False).head(10),0)
```

**Insights:** "AMAZON FEE" description sales more compare to other Description in the year 2010.

```python
round(df1[df1.Year==2011].groupby("Description")["Sales"].sum().sort_values(ascending=False).head(10),0)
```

**Insights:** "PAPER CRAFT , LITTLE BIRDIE" description sales more compare to other Description in the year 2011.

```python
df1[df1.Country=="United Kingdom"].groupby("Year-Month")["Sales"].sum().plot()
```

**Insights:** "United Kingdom" has more sales in the Month of November in 2011.

```python
df1[df1.Year==2010].groupby("Month")["Sales"].sum().sort_values(ascending=True)
```

```python
df1[df1.Year==2011].groupby("Month")["Sales"].sum().sort_values(ascending=False)
```

```python
df1[df1.Country=="Germany"].groupby("Year-Month")["Sales"].sum().plot()
```

**Insights:** Germany has more sales in October month in 2011.
